```python
verbose = True
ans = 0
ans_2 = 0

"""
So first try was round. built in python func. but
there is no auto round down mode to it... why they
wouldnt just use math.floor to return an easy
round_down function... maybe i should pep about it.
Anyway, a quick google shows math.floor will round
you down, no worries.
"""
def calc(mass): return math.floor(mass /3) -2
    
for mass in list__:
    add_to_ans = 0
    fuel_for_mass = calc(mass)
    add_to_ans += fuel_for_mass
    ans += fuel_for_mass
    if verbose: print(f"Takes {fuel_for_mass} fuel, to carry {mass} mass")
    while True:
        last = fuel_for_mass
        fuel_for_mass = calc(fuel_for_mass)
        if fuel_for_mass > 0:
            if verbose: print(f"Takes {fuel_for_mass} fuel to carry {last} amount of fuel")
            add_to_ans += fuel_for_mass
        else:
            break
    ans_2 += add_to_ans
print(f"Part 1 Answer: {ans}")
print(f"Part 2 Answer: {ans_2}")
```
